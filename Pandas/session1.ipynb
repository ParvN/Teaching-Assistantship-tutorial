#### How to add and delete rows in existing dataframe?

```python
data.loc[len(data.index)+1] = [len(data.index)+1, 5, 3.6, 1.4, 0.2, 'Iris-setosa']
data.tail(3)
```

```python
new_row1 = {'Id': len(data.index) + 1, 'SepalLengthCm': 4.6, 'SepalWidthCm': 3.1, 'PetalLengthCm': 1.5, 'PetalWidthCm': 0.2, 'Species': 'Iris-setosa'}
data = data.append(new_row1, ignore_index=True)
data.tail(3)
```

```python
data = pd.concat([data, pd.DataFrame([new_row1])], ignore_index = True) #dictionary needs to be converted to dataframe first for concat
data.tail(3)
```

```python
data = data.drop([data.index[-1], data.index[-2]])
data.tail(3)
```

How to add and delete columns in dataframe?

```python
#Insert as last column
import numpy as np
data.insert(len(data.columns), "Test_column1", np.zeros((data.shape[0])))
data.head()
```

```python
data = data.assign(Test_column2 = np.zeros((data.shape[0])))
data.head()
```

```python
data.drop(["Test_column1", "Test_column2"], axis = 1, inplace = True)
data.head()
```

#### Renaming columns

```python
data.rename(columns = {'Species': 'species'}, inplace = True)
data.head()
data.rename(columns = {'species': 'Species'}, inplace = True)
```
